```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8')
%matplotlib inline
```

## 1. Load Data

```python
# Load the dataset
df = pd.read_csv('../data/heart.csv')

print(f"Dataset shape: {df.shape}")
print(f"\nColumns: {df.columns.tolist()}")
print(f"\nFirst few rows:")
df.head()
```

## 2. Basic Statistics

```python
# Basic info
print("\n=== Dataset Info ===")
df.info()

print("\n=== Statistical Summary ===")
df.describe()
```

## 3. Missing Values Check

```python
print("\n=== Missing Values ===")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])

if missing_values.sum() == 0:
    print("No missing values found!")
```

## 4. Target Variable Distribution (Class Balance)

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='target', data=df, palette='Set2')
plt.title('Class Distribution: Heart Disease (1) vs No Heart Disease (0)', fontsize=14)
plt.xlabel('Target')
plt.ylabel('Count')

# Add count labels
for i in range(df['target'].nunique()):
    count = df['target'].value_counts()[i]
    percentage = (count / len(df)) * 100
    plt.text(i, count, f'{count}\n({percentage:.1f}%)', ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.savefig('../reports/class_distribution.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"Class balance:")
print(df['target'].value_counts())
```

## 5. Numerical Feature Distributions

```python
# Numerical columns
numerical_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

for idx, col in enumerate(numerical_cols):
    if idx < len(axes):
        axes[idx].hist(df[col], bins=20, color='skyblue', edgecolor='black')
        axes[idx].set_title(f'Distribution of {col}', fontsize=12)
        axes[idx].set_xlabel(col)
        axes[idx].set_ylabel('Frequency')

# Remove empty subplot
if len(numerical_cols) < len(axes):
    for idx in range(len(numerical_cols), len(axes)):
        fig.delaxes(axes[idx])

plt.tight_layout()
plt.savefig('../reports/numerical_distributions.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 6. Correlation Heatmap

```python
plt.figure(figsize=(12, 10))

# Calculate correlation matrix
correlation_matrix = df.corr()

# Create heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8}, fmt='.2f')

plt.title('Feature Correlation Heatmap', fontsize=16, pad=20)
plt.tight_layout()
plt.savefig('../reports/correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nTop correlations with target:")
target_corr = correlation_matrix['target'].sort_values(ascending=False)
print(target_corr)
```

## 7. Categorical Feature Analysis

```python
# Categorical columns
categorical_cols = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']

fig, axes = plt.subplots(2, 4, figsize=(18, 10))
axes = axes.ravel()

for idx, col in enumerate(categorical_cols):
    if idx < len(axes):
        sns.countplot(x=col, data=df, hue='target', ax=axes[idx], palette='Set1')
        axes[idx].set_title(f'{col} vs Target', fontsize=11)
        axes[idx].legend(title='Target', fontsize=8)

plt.tight_layout()
plt.savefig('../reports/categorical_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 8. Age Distribution by Target

```python
plt.figure(figsize=(10, 6))

sns.boxplot(x='target', y='age', data=df, palette='Set2')
plt.title('Age Distribution by Heart Disease Status', fontsize=14)
plt.xlabel('Heart Disease (0=No, 1=Yes)')
plt.ylabel('Age')

plt.tight_layout()
plt.savefig('../reports/age_by_target.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nAge statistics by target:")
print(df.groupby('target')['age'].describe())
```

## 9. Key Insights Summary

### Observations from EDA:

1. **Class Balance**: The dataset is reasonably balanced with approximately 54% having heart disease and 46% not having it.

2. **Age Distribution**: Most patients are between 40-65 years old. Age shows some correlation with heart disease risk.

3. **Key Correlations**: 
   - Positive correlation with target: cp (chest pain type), thalach (max heart rate), slope
   - Negative correlation with target: oldpeak, ca, thal, exang

4. **Missing Values**: No missing values in the cleaned dataset.

5. **Feature Selection**: Based on correlation analysis, features like cp, thalach, oldpeak, ca, and thal appear most predictive.
